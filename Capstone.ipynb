```python
#find mean percentile for each domain
gb= df.groupby("Domain").agg({"Percentile 90th": 'mean', "Percentile 75th": 'mean', "Percentile 50th": 'mean', "Percentile 25th": 'mean'})
print(gb)
```

```python
#create heatmap visualization for easier the darker the shade, the lower the mark, the lighter the shade the higher mark
plt.figure(figsize=(10,6))
plt.title("Average Percentile by Domain")
sns.heatmap(gb,annot = True)
plt.show()
```

```python
#from the chart above we can see that satisfaction is highest in the Acess to primary care Domain throughout all plans and that Use of services has the lowest
```

```python
#find out average rate for each plan
gb1=df.groupby("Plan Name").agg({"Rate": 'mean'})
print(gb1)
plt.figure(figsize=(10,15))
plt.title("Average Rate by Plan Name")
sns.heatmap(gb1, annot = True)
plt.show()
```
