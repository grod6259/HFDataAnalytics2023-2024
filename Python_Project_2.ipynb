I wanted to find out if there was a coorelation between Scores and number of students who took the test. Filtering the data to remove any null values, further filtering to focus on the Highschool level and then finding out what Highschool has the highest and lowest mean scores were just for reference.

In the entire data set it seems the more students take the test the higher the mean score.

In the Highschool with the lowest mean scores, it seems to show the opposite, there is a slight downward trend, the more students test, the lower the scores

In the Highschool with the highest mean score, the more students tested the least amount of students mean scored higher than 80.
